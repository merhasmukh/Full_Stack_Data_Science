## Text Lowercase

We do lowercase the text to reduce the size of the vocabulary of our text data

For any algorithm "Hasmukh Mer" and "hasmukh mer" are two different things so it creates two vector value for both.

So it;s good to have all the text in the same case lower or upper.

```python
def lower_text(text):
    return text.lower()

input_str="Hasmukh Mer"
lower_text(input_str)
```

## Remove numbers

We should either remove the numbers or convert those numbers into textual representations .

We use regular expressions(re) to remove the numbers.



```python
#For Removing numbers 

def remove_num(text):
    result=re.sub('\d+',"",text)
    return result

input_s="Hasmukh 56 Mer"
remove_num(input_s)
```

```python
import re
def find_gmail(text):
    return re.findall("\S+@\S+",text)

h="hasmukh mer merhasmukh20@gmail.com sarva botad mer@hasmukh "

find_gmail(h)
```

Here you can see that mer@hasmukh is not a gmail but our regular expression find it as a gmail so we have to change our regular expression.



```python
import re
def find_gmail(text):
    return re.findall("[\w\.\+\-]+\@[\w]+\.[a-z]{2,3}",text)

h="hasmukh mer merhasmukh20@gmail.com sarva botad mer@hasmukh "

find_gmail(h)
```

```python
#import the library
import inflect

q= inflect.engine()


# convert number into text

def convert_num(text):
    #split strings into list of texts 
    temp_string=text.split()
    
    #initialise empty list
    new_str=[]
    print(temp_string)
    
    for word in temp_string:
        #if text is a digit, convert the digit
        # to numbers and append into the new_str list
        
        if word.isdigit():
            temp=q.number_to_words(word,andword="")
            new_str.append(temp)
        
        #append the text as it is
        else:
            new_str.append(word)
            
            
    # join the texts of new_str to from a string
    temp_str=" ".join(new_str)
    
    return temp_str


input_str="hasmukh mer 57"

convert_num(input_str)
    
```

## Remove Punctation

We remove punctuation because of that we don't  have different form of the same word. If we don't remove punctuation, then been,been, and been! will be treated separately. 

```python
# Let's remove punctuation
import string
def rem_punctuation(text):
    translator=str.maketrans("","",string.punctuation)
    return text.translate(translator)

input_str="Hey !!  ,Hasmukh Mer?? ,Are you okay??"

rem_punctuation(input_str)
```

In some cases we don't need punctuatios so we can remove it. as an exaple if we want to do sentiment analysis then we don't need punctuations marks so we simply remove it in this case.

In other case like text generator we don't need to remove punctuations because we need to generate punctuations as partof our text generator.

So, It depends on the task on what we apply text preprocessing.
